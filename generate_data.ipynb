```python
from faker import Faker
import pandas as pd
from datetime import datetime
import dataframe_image as dfi
```

```python
fake = Faker("ru_RU")
```

```python
n_samples = 4000
```

```python
names = [fake.name() for _ in range(int(0.9 * n_samples))]
names.extend([fake.first_name() + ' ' + fake.last_name() for _ in range(int(0.1 * n_samples))])

phone_numbers = [fake.phone_number() for _ in range(n_samples)]

emails = [fake.ascii_email() for _ in range(n_samples)]

addresses = [fake.address() for _ in range(n_samples)]

credit_cards = [fake.credit_card_number() for _ in range(n_samples)]

for i in range(int(0.8 * n_samples), n_samples):
    num = credit_cards[i]
    credit_cards[i] = num[:4] + ' ' + num[4:8] + ' ' + num[8:12] + ' ' + num[12:]

security_codes = [fake.credit_card_security_code() for _ in range(n_samples)]

dates = [fake.date(pattern = '%Y-%m-%d', end_datetime = datetime.strptime('2016-05-24', '%Y-%m-%d')) 
               for _ in range(int(0.8 * n_samples))]
dates.extend([fake.date(pattern = '%Y/%m/%d', end_datetime = datetime.strptime('2016/05/24', '%Y/%m/%d')) 
               for _ in range(int(0.1 * n_samples))])
dates.extend([fake.date(pattern = '%B %d, %Y', end_datetime = datetime.strptime('May 24, 2016', '%B %d, %Y')) 
               for _ in range(int(0.1 * n_samples))])
```

```python
pii_df = pd.DataFrame({'name': names, 
                       'phone_number': phone_numbers, 
                       'email': emails,
                       'address': addresses, 
                       'credit_card': credit_cards, 
                       'security_code': security_codes, 
                       'date': dates})
```
